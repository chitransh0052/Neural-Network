```python
# prompt: make scatter plot of this

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Assuming X and y are already defined from the previous code block
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=1,
                           n_redundant=0, n_clusters_per_class=1, hypercube=False,
                           random_state=43, class_sep=15)

# Create the scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot of Data Points")
plt.show()
```

```python
def perceptron(X,y):
    w1=w2=b=1
    eta=0.1
    for i in range(1000):
      for j in range(100):
        z=w1*X[j][0]+w2*X[j][1]+b
        if y[j]*z<0:
          w1=w1+eta*y[i]*X[j][0]
          w2=w2+eta*y[i]*X[j][1]
          b=b+eta*y[i]
    return w1,w2,b
w1,w2,b=perceptron(X,y)
```

```python
print("w1:",w1)
print("w2:",w2)
print("b:",b)
```

```python
m=(-w1/w2)
c=(-b/w2)
```

```python
import numpy as np
```

```python
x=np.linspace(-3,3,100)
Y=m*x+c
plt.plot(x,Y)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,3)
```
